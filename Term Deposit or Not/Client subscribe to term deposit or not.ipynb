#### Replacing nan values with most frequent category for columns "job", "marital", "education", "housing", "loan"

```python
def impute_most_frequent(df, variable):
    most_frequent = df[variable].mode()[0]
    df[variable] = df[variable].fillna(most_frequent)
```

```python
for column in ["job", "marital", "education", "housing", "loan"]:
    impute_most_frequent(df, column)
```

```python
df.isnull().sum()
```

#### Since default have more number of missing values i will replace the NaN values with word "missing"

```python
df["default"]=df["default"].fillna("missing")
```

```python
df.isnull().sum()
```

#### Splitting data into train and test

```python
X = df.drop(columns = ["y"])
y = df["y"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.80, random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.columns
```

```python
cat_columns = X_train.select_dtypes(include = ["object"]).columns
print(cat_columns)
```

```python
num_columns = X_train.select_dtypes(include = ["number"]).columns
print(num_columns)
```

#### Applying label encoder

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
for column_name in cat_columns:
    label_encoder.fit(X_train[column_name])
    X_train[column_name] = label_encoder.transform(X_train[column_name])
    X_test[column_name] = label_encoder.transform(X_test[column_name])
```

```python
X_train
```

```python
X_test
```

#### Selecting important features based on random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
features = []
importances = []
for feature, importance in zip(X_train.columns, clf.feature_importances_):
    features.append(feature)
    importances.append(importance)
```

```python
feature_importances = pd.DataFrame({"Features":features,
                                    "Importances":importances})
```

```python
feature_importances
```

```python
plt.bar(feature_importances["Features"],feature_importances["Importances"])
plt.xticks(rotation = "vertical")
plt.show()
```

```python
feature_importances.sort_values(by = "Importances", ascending = False)
```

#### So iam selecting the features which are more than 2% importance

```python
feature_importances[feature_importances["Importances"] >= 0.02]
```

```python
selected_features = feature_importances[feature_importances["Importances"] >= 0.02]["Features"].to_list()
print(selected_features)
```

#### Updating X_train and X_test with selected features

```python
X_train = X_train[selected_features]
X_train.head()
```

```python
X_test = X_test[selected_features]
X_test.head()
```

#### Handling Unbalanced Data using SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE()
```

```python
X_train, y_train = oversample.fit_resample(X_train, y_train)
```

```python
y_train.value_counts()
```

#### Now the data is balanced..............................

#### Applying Standard scaler to numerical columns

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
```

```python
SS.fit(X_train)
```

```python
X_train = pd.DataFrame(SS.transform(X_train), columns = X_train.columns)
X_train.head()
```

```python
X_test = pd.DataFrame(SS.transform(X_test), columns = X_train.columns)
X_test.head()
```

#### Now all the preprocessing steps are done.......................

### Model Building

#### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
Log_reg = LogisticRegression()
```

```python
Log_reg.fit(X_train, y_train)
```

```python
y_pred = Log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score
```

```python
print(classification_report(y_test, y_pred))
print("--------------------------------------------------------")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
```

#### 2. Adaboost 

```python
from sklearn.ensemble import AdaBoostClassifier

abc = AdaBoostClassifier()

abc.fit(X_train, y_train)

abc_pred = abc.predict(X_test)
```

```python
print(classification_report(y_test, abc_pred))
print("----------------------------")
ps = precision_score(y_test, abc_pred)
rc = recall_score(y_test, abc_pred)
acc = accuracy_score(y_test, abc_pred)
print("Accuracy:", acc)
print("Precision:", ps)
print("Recall:", rc)
```
